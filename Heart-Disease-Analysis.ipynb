```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
sns.displot(data=df, x=df['Age'],color='blue')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
df['Sex'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.subplot(1,3,2)
df['HeartDisease'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.subplot(1,3,3)
df['ExerciseAngina'].value_counts().plot(kind='pie',autopct='%1.1f%%')

plt.show()
```

```python
ax = sns.countplot(data=df,x=df['HeartDisease'], hue='Sex')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.head(3)
```

```python
# ExerciseAngina - pain in the chest that comes on with exercise, stress, or other things that make the heart work harder
ax = sns.countplot(data=df,x=df['ExerciseAngina'], hue='HeartDisease')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sns.boxplot(data=df, x=df['RestingBP'], color='blue')
```

```python
df['RestingBP'].unique()
```

```python
df = df[df['RestingBP'] > 75]
```

```python
df.head(3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 7))

sns.histplot(ax=axes[0], data=df, x=df['RestingBP'],kde=True,color='blue')

sns.histplot(ax=axes[1], data=df, x=df['Cholesterol'],kde=True, color='blue')

sns.histplot(ax=axes[2], data=df, x=df['MaxHR'],kde=True, color='blue')

sns.histplot(ax=axes[3], data=df,x=df['Oldpeak'],color='blue')

plt.show()
```

```python
group_data = df.groupby(['Sex','ExerciseAngina'])['HeartDisease'].value_counts()
group_data 
```

```python
group_data.plot(kind='bar',color='blue')
plt.title('Heart Disease vs. Exercise Angina by Gender')
plt.xlabel('Gender and ExerciseAngina')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
